# Introduction

## Background

The advent of online hotel reservation platforms has revolutionized the way bookings are made and influenced customer behavior significantly. A considerable portion of hotel reservations is canceled due to various reasons such as changes in plans or scheduling conflicts. The availability of free or low-cost cancellation options has made this process more convenient for guests, but it poses challenges for hotels as it can lead to decreased revenue and is generally less favorable for them to manage.

## Key Terms

- **Dataframe** is two-dimensional, size-mutable, potentially heterogeneous tabular data. (Source: https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.html)
- **Dummy variable** takes a binary value (0 or 1) to indicate the absence or presence of some categorical effect that may be expected to shift the outcome.(Source: Draper, N.R.; Smith, H. (1998) Applied Regression Analysis, Wiley. ISBN 0-471-17082-8 (Chapter 14))
- **Train Test Split** is a model validation process that allows you to simulate how your model would perform with new data. (https://builtin.com/data-science/train-test-split)
- **Random Tree Classifier** is a meta-estimator that builds multiple decision tree classifiers on different subsets of the dataset and employs averaging to enhance predictive accuracy while managing overfitting. (https://scikit-learn.org/stable/modules/generated/sklearn.ensemble.RandomForestClassifier.html)
- **Decision Trees (DTs)** are an unsupervised learning technique utilized in both classification and regression tasks. Their aim is to construct a predictive model that determines the target variable's value by understanding straightforward decision rules derived from the features within the data. Essentially, a tree represents a segmented, constant-value approximation. (https://scikit-learn.org/stable/modules/tree.html#tree)

## Research Question

This project is designed to explore the possibility to predict if the customer is going to honor the reservation or cancel it based on factors such as types of booking and customers.

# Data Description

The dataset is gained from Kaggle with the link: https://www.kaggle.com/datasets/ahsan81/hotel-reservations-classification-dataset/data and which is inspired by the rseearch (https://www.sciencedirect.com/science/article/pii/S2352340918315191). The dataset contains hotel reservation data collected from two hotels in Portugal. It includes the following fields:

- **Booking_ID**: unique identifier of each booking
- **no_of_adults**: Number of adults
- **no_of_children**: Number of Children
- **no_of_weekend_nights**: Number of weekend nights (Saturday or Sunday) the guest stayed or booked to stay at the hotel
- **no_of_week_nights**: Number of week nights (Monday to Friday) the guest stayed or booked to stay at the hotel
- **type_of_meal_plan**: Type of meal plan booked by the customer:
    - Not Selected – No meal plan selected
    - Meal Plan 1 – Breakfast
    - Meal Plan 2 – Half board (breakfast and one other meal)
    - Meal Plan 3 – Full board (breakfast, lunch, and dinner)
- **required_car_parking_space**: Does the customer require a car parking space? (0 - No, 1- Yes)
- **room_type_reserved**: Type of room reserved by the customer. The values are ciphered (encoded) by INN Hotels.
- **lead_time**: Number of days between the date of booking and the arrival date
- **arrival_year**: Year of arrival date
- **arrival_month**: Month of arrival date
- **arrival_date**: Date of the month
- **market_segment_type**: Market segment designation.
- **repeated_guest**: Is the customer a repeated guest? (0 - No, 1- Yes)
- **no_of_previous_cancellations**: Number of previous bookings that were canceled by the customer prior to the current booking
- **no_of_previous_bookings_not_canceled**: Number of previous bookings not canceled by the customer prior to the current booking
- **avg_price_per_room**: Average price per day of the reservation; prices of the rooms are dynamic. (in euros)
- **no_of_special_requests**: Total number of special requests made by the customer (e.g. high floor, view from the room, etc)
- **booking_status**: Flag indicating if the booking was canceled or not.

# Exploratory Data Analysis and Processing

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
from sklearn import cluster
from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import GridSearchCV
from sklearn.tree import DecisionTreeClassifier
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/stevenzhouzh/info1998-final/main/Hotel%20Reservations.csv")
df.head()
```

```python
df.info()
```

```python
df.columns
```

We continue to explore the statistical data behind each field. From the data below we can see a mean number of adult customers of 1.8 which is a reasonable number because most customers will choose rooms for one person or two persons. Also, there is a significantly lower number of children with a mean number of 0.8 highlighting that adult customers will be the primary research target (less focus on family hotel reservations). 

```python
df.describe()
```

Take columns related to arrival dates ('arrival_year', 'arrival_month', 'arrival_date'), converts them into a single datetime column ('arrival_date'), and then removes the original columns ('arrival_year' and 'arrival_month') from the DataFrame.

```python
df['arrival_date']= pd.to_datetime(df[['arrival_year', 'arrival_month', 'arrival_date']].\
                                   rename({'arrival_year':'year', 'arrival_month':'month', 'arrival_date':'day'}, axis=1), 
                                   errors='coerce')
df = df.drop(['arrival_year', 'arrival_month'], axis=1)
```

Separately storing columns containing numerical data into `numeric_cols` and categorical data into `cat_cols` to streamline the following steps to visualize data distribution and dummify categorical features.

```python
numeric_cols = ['no_of_adults', 'no_of_children', 'no_of_weekend_nights', 'no_of_week_nights', 'lead_time', 
                'no_of_previous_cancellations', 'no_of_previous_bookings_not_canceled',
                'avg_price_per_room', 'no_of_special_requests', 'arrival_date']
cat_cols = ['type_of_meal_plan', 'required_car_parking_space', 'room_type_reserved', 'market_segment_type', 'repeated_guest', 'booking_status']
```

When we examine these features, we found it interesting to see the distribution of lead time (which shows the number of days between the date of booking and the arrival date) to examine the customers' booking behavior which could influence their later decision in cancellation.

We use histogram to visualize the lead time between booking and arrivial and find that the distribution is right-skewed. We can infer that many customers prefer to book their hotels shortly before their planned arrival dates. This could be due to various reasons such as last-minute planning, flexible schedules, or simply a preference for making arrangements closer to the actual travel date.

```python
sns.histplot(data=df, x="lead_time")

# Set labels and title
plt.xlabel("Lead Time")
plt.ylabel("Frequency")
plt.title("Distribution of Lead Time between Booking and Arrival")

# Show the plot
plt.show()
```

In a larger scale, we examine the number of bookings across 1 year from 2017 to 2018 and find out that there tend to be more bookings in 2018 than in 2017.

```python
sns.histplot(df, x='arrival_date')
plt.xticks(rotation=45)
plt.show()
```

Distributions for other numerical features are also presented as below.

```python
for var in numeric_cols:
    ax = sns.histplot(df, x=var)
    plt.show()
```

We use a bar graph to examine the booking status, we can see that the canceled reservation is almost half of the not canceled reservation. 

```python
sns.countplot(data=df, x="booking_status")

# Set labels and title
plt.xlabel("Booking Status")
plt.ylabel("Count")
plt.title("Distribution of Booking Status")

# Show the plot
plt.show()
```

Distributions of other categorical features are presented as below:

```python
for var in cat_cols:
    ax = sns.histplot(df, x=var)
    plt.show()
```

Since many machine learning algorithms require numerical input data, categorical variables need to be transformed into a numerical format for these algorithms to work effectively. After processing with the following code, dummies will contain the dummified representation of categorical features, which can then be used in machine learning models that require numerical input data. Each column in dummies corresponds to a category from the original categorical variables, and each row represents a data point with binary indicators for the presence or absence of each category.

```python
dummies = pd.DataFrame()
for cat_var in cat_cols:
    dummies = pd.concat([dummies, pd.get_dummies(df[cat_var], drop_first=True)], axis=1)
dummies.head()
```

Then, we concatenate the original DataFrame df with the dummified DataFrame dummies, and drops the original categorical columns (cat_cols) from df_dummified.

```python
df_dummified = pd.concat([df, dummies], axis=1).drop(cat_cols, axis=1)
df_dummified = df_dummified.drop([1], axis=1)
print(df_dummified.head())
```

Since there are too many features, we decide to computes the correlation matrix for the dummified DataFrame and see what features have high correlation.

```python
corr = df_dummified.drop(['Booking_ID'], axis=1).corr()
corr
```

To better visualize it, we use a heatmap to show the correlation matrix. Darker red represents higher positive correlation and darker blue represents higher negative correlation.

```python
plt.figure(figsize=(10, 8))
sns.heatmap(corr, annot=False, cmap='coolwarm', fmt=".2f")
plt.title('Heatmap of the Correlation Matrix')
plt.show()
```

Besides the strongest positive correlation between the variable and itself and the strongest negative correlation between the online and offline booking, we noticed that there is a negative correlation between the lead time and the non-canceled reservation. This could be due to several reasons. Customers who book closer to their arrival date might have more concrete plans and are less likely to cancel compared to those who book far in advance. Also, bookings made with shorter lead times may indicate more flexible travel plans or business trips, which are less likely to be canceled.

We visualize the proportion of cancelation for different lead time and find that ther is higher proportion of cancelation found with longer lead time (shown as more blue section than orange section in the graph). 

```python
sns.histplot(data = df_dummified, x= 'lead_time', hue = 'Not_Canceled', multiple = 'stack')
plt.show()
```

Potential reason is that customers who book well in advance may encounter changes in their plans over time. These changes could be due to various reasons such as unexpected events, travel restrictions, or alterations in personal schedules. As a result, they may be more likely to cancel their bookings compared to those who book closer to their arrival date and have more immediate plans.

# Machine Learning Models

We use machine learning model to predict the booking status to be cancelled or not cancelled.

To better evaluate and compare the performance of different machine learning models, a list is used to store the accuracy score of each machine learning models (after optimization).

```python
accuracy_list = []
models = ["Random Forest", "Decision Tree"]
```

## Random Forest

```python
data = df_dummified
# Define features and target variable
features = list(data.select_dtypes(include=[np.number]).columns.values)
print(f'Features: {features}')
# Select all columns except the last one as features
target = 'Not_Canceled'

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(data[features], data[target], test_size=0.2, random_state=42)

# Initialize Random Forest classifier
rf_classifier = RandomForestClassifier(n_estimators=500, random_state=42)

# Train the classifier
rf_classifier.fit(X_train, y_train)

# Make predictions
y_pred = rf_classifier.predict(X_test)

# Evaluate accuracy
original_random_forest_accuracy_score = accuracy_score(y_test, y_pred)
print(f'Accuracy: {original_random_forest_accuracy_score}')
```

### Optimization

Tune the hyperparameters of the Random Forest model to find the best combination for your data. Parameters like n_estimators, max_depth, min_samples_split, and min_samples_leaf can significantly impact the model's accuracy.

```python
# Define the parameter grid for hyperparameter tuning
param_grid = {
    'n_estimators': [100, 200, 300],
    'max_depth': [None, 10, 20],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}

# Initialize GridSearchCV to find the best hyperparameters
grid_search = GridSearchCV(rf_classifier, param_grid, cv=5, scoring='accuracy')

# Perform hyperparameter tuning
grid_search.fit(X_train, y_train)

# Get the best hyperparameters
best_params = grid_search.best_params_
```

```python
# The best hyperparameters
print(best_params)
```

```python
best_rf_classifier = RandomForestClassifier(
    n_estimators=300,
    max_depth=20,
    min_samples_split=2,
    min_samples_leaf=1,
    random_state=42
)

# Train the classifier with the full training data (X_train, y_train)
best_rf_classifier.fit(X_train, y_train)

# Make predictions using the trained classifier
y_pred_best = best_rf_classifier.predict(X_test)

# Evaluate accuracy using the best model
best_random_forest_accuracy_score = accuracy_score(y_test, y_pred_best)
print(f'Accuracy with best hyperparameters: {best_random_forest_accuracy_score}')

random_forest_accuracy_increase = ((best_random_forest_accuracy_score - original_random_forest_accuracy_score) / original_random_forest_accuracy_score) * 100
# Round the percentage increase to three decimal places
random_forest_accuracy_increase_rounded = round(random_forest_accuracy_increase, 3)

print(f"The accuracy has increased by {random_forest_accuracy_increase_rounded}%")

accuracy_list.append(best_random_forest_accuracy_score)
```

## Decision Tree

```python
tree = DecisionTreeClassifier(max_depth = 1)
tree.fit(X_train, y_train)

# Store the predictions for the training and test set
pred_train = tree.predict(X_train)
pred_test = tree.predict(X_test)

original_decision_tree_accuracy_score = accuracy_score(y_test, pred_test)
# Print the test and training accuracy
print("Test Accuracy: ", original_decision_tree_accuracy_score)
```

### Optimization

```python
# Hyperparameter tuning using GridSearchCV
param_grid = {
    'max_depth': [1, 5, 10, None], 
    'min_samples_split': [2, 5, 10], 
    'min_samples_leaf': [1, 2, 4] 
}

# Initialize GridSearchCV for hyperparameter tuning
grid_search = GridSearchCV(DecisionTreeClassifier(), param_grid, cv=5, scoring='accuracy')
grid_search.fit(X_train, y_train)

# Get the best hyperparameters
best_params = grid_search.best_params_
print("Best Hyperparameters:", best_params)

# Initialize Decision Tree classifier with best hyperparameters
best_tree = DecisionTreeClassifier(**best_params)
best_tree.fit(X_train, y_train)

# Store the predictions for the training and test set using the best model
best_pred_train = best_tree.predict(X_train)
best_pred_test = best_tree.predict(X_test)

# Print the test and training accuracy using the best model
best_decision_tree_accuracy_score = accuracy_score(y_test, best_pred_test)
print("Test Accuracy (Best Decision Tree):", best_decision_tree_accuracy_score)

decision_tree_accuracy_increase = ((best_decision_tree_accuracy_score - original_decision_tree_accuracy_score) / original_decision_tree_accuracy_score) * 100

# Round the percentage increase to three decimal places
decision_tree_accuracy_increase_rounded = round(decision_tree_accuracy_increase, 3)

print(f"The accuracy has increased by {decision_tree_accuracy_increase_rounded}%")

accuracy_list.append(best_decision_tree_accuracy_score)
```

## Performance of Machine Learning Models

```python
ax = sns.barplot(x=models, y=accuracy_list)
for i in ax.containers:
    ax.bar_label(i,)
plt.xlabel("Classifiers", fontsize=15)
plt.ylabel("Accuracy (%)", fontsize=15)
plt.title("List of Accuracies")
plt.show()
accuracy_list
```
